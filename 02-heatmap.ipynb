```python
# %pip install folium
import pandas as pd
import folium
from folium.plugins import HeatMap

df = pd.read_csv('./datasets/shooting.csv', low_memory=False)
```

```python
df = df[pd.notnull(df['Latitude'])]
df = df[pd.notnull(df['Longitude'])]
df['Count'] = 1
```

```python
def generateBaseMap(default_location = [df['Latitude'].median(), 
                                        df['Longitude'].median()], 
                    default_zoom_start = 12):
    
    return folium.Map(location = default_location, control_scale = True, zoom_start = default_zoom_start)
```

```python
base_map = generateBaseMap()

HeatMap(data = df[['Latitude', 'Longitude', 'Count']]
            .groupby(['Latitude', 'Longitude'])
            .sum()
            .reset_index()
            .values.tolist(),
        radius = 8,
        max_zoom = 11).add_to(base_map)

display(base_map)
```
